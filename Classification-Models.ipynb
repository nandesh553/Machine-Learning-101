### Classification Models

Classification is used to classify data in diffrent categories. They are also known as classifiers. Classification model gives a probality score and checks it with certain threshold to classify labels.

* True or False # binary classifier
* Up or Down # binary classifier
* Low, Medium or High

#### Types of classificatiion
* Binary Classifier : Only two output variables or labels. Eg (True or False), (Up or Down)
* Multi-label Classifier : Ouput is tuple of multiple binary variables
* Multi-class Classifier : Output variables(N) are greater than 2 => N > 2. Eg Digit Classification from 0-9 or A-Z
* Multi-output Classifier : multi-label + multi-class

#### Metrics used for evaluation
* Accuracy
* Precision
* ROC

#### Models
* Logistic Regression - [binary]
* Decision Tree
* Naive Bayes - [multi-label]
* Support Vector Machines (SVM) - [binary]
* Nearest Neighbours

#### Loss Function / Cost Function
It is used to calculate the loss and then used to further improve the model to create it robust .
* Cross Entropy

#### Optimization
* Hyperparameter Tuning
* Grid Search

Binary classifiers can be generalised for multi-label classification

Lets say we need to classify digits from 0-9

#### One vs All
We train 10 diffrent binary classifiers for all the digits and then choose the one with highest probability

#### One vs One
Train 45 binary classifier for each combination
* One detector for each pair of digits
* For N labels, need $\frac{N(N-1)}{2} classifiers

## You have a fully trained classififcation model but how do you know that it is a good one? 

We do it using metrics to eavluate our model. The metrics used to evaluate classification models are:
* Accuracy
* Precision
* Recall

Lets understand some terms first
* TP = True Positives = Rightly predicted positives
* TN = True Negatives = Rightly predicted negatives
* FP = False Positives = Falsly predicted negatives
* FN = False Negatives = Falsly predicted negatives
* Toatl Number of instances / labels

#### Accuracy
Predicted Labels vs Actual Labels

$$ \frac{TP + TN}{ Number Of instances } $$

More matches = High Accuracy

High accuracy is good but you need to understand an algorithm might have high accuracy but **still be poor** machine learning model. Its predictions are useless.

For eg. there is a model to predict cancer in a patient. Actual probabilty of a person having this disease is very rare. Like a 1 in 10000 have this disease. And if our model's accuracy is 99.999% then **still it is bad**. Because the error rate will give false for person having the disease.

Accuracy does not work on <a href="https://en.wikipedia.org/wiki/Skewness">skewed</a> dataset. When some labels are more common then other then accuracy is a poor evaluation metric. This is where confusion matrix comes in handy.


#### Confusion matrix

Confusion matrix is simpoly a grid tabulating the predicted label from the model vs the actual model. Predicted Labels are among the roews and Actual Labels are among the rows.


#### Precision
$$ \frac{TP}{TP + FP} $$

precison decides conservativeness of classifier. 

#### Recall

$$ \frac{TP}{TP + FN} $$

#### Precision and Recall are inversely proportional. So hihger precision means lower recall and vice versa.

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.subplots(figsize=(20,10))

x = np.linspace(-10, 10, 100)
y_precision = 1 / (1 + np.exp(-x))
y_recall = 1 / (1 + np.exp(x))

plt.plot(x, y_precision)
plt.plot(x, y_recall)

plt.axvline(-5)

plt.legend(['precision', 'recall'])

plt.xlabel('"Conservativeness" of classifier', fontsize=18)

plt.show()
```

As we can see from the graph above => **Precision and Recall are inversely proportional.** So hihger precision means lower recall and vice versa.

If this is the case **How do we find a model that is good.** There are several heuristics available one of them is **ROC Curve**.

#### Receiver Operating Characteristic

```python
fig = plt.subplots(figsize=(20,10))

plt.xlabel('False Positive Rate', fontsize=18)
plt.ylabel('True Positive Rate', fontsize=18)

plt.show()

#### TO-DO ROC Curve
```
